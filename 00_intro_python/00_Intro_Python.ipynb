## ¿Qué es una celda en Jupyter Notebook?

Un **notebook** está compuesto por unidades llamadas **celdas**. Existen principalmente dos tipos de celdas:

### 1. Celdas de código

Estas celdas permiten escribir y ejecutar código Python. Al ejecutar una celda de código (por ejemplo, presionando `Shift + Enter` (MAC o Linux) o `Ctr + Enter` (Windows)), el resultado se mostrará justo debajo de la celda.

```python
# Esto es una celda de código
print("Hola, mundo")
```

Puedes modificar el código y volver a ejecutar la celda tantas veces como necesites. Esto hace que sea un entorno muy útil para experimentar, probar y explorar ideas rápidamente. 

**NOTA**: Es importante saber que el orden en que ejecutes las celdas sí importa. Cada vez que ejecutas una celda, su contenido se ejecuta en el contexto actual de Python, que se mantiene activo en la memoria del kernel.

Por ejemplo, si ejecutamos primeramente la siguiente celda de código:

```python
x = 5
```

entonces, luego puedes usar esta variable en otra celda, 

```python
print(x)
```

incluso si está celda se encuentra más arriba o más abajo en el notebook. El código de esta última celda funcionará siempre y cuando hayas ejecutado primero la celda donde se definión la variable `x`. Si ejecutas una celda que depende de otra que aún no se ha ejecutado, recibirás un error. Por favor toma en cuenta esto y escribe tu código de una forma ordenada. 

**RECOMENDACIÓN**: Si no estás seguro del orden de ejecución, puedes ir al menú y seleccionar `Kernel > Restart` para reiniciar el kernel (es decir, limpiar la memoria) y volver a ejecutar las celdas en un orden adecuado.

### 2. Celdas de Markdown

La celda de tipo **Markdown** (como en la que estoy escribiendo este texto) nos permiten escribir texto con formato (como el que utilizan los blogs, por ejemplo), incluir títulos, listas, fórmulas mátemáticas (usando LaTeX), imágenes, enlaces y más. No se ejecutan como código, sino que se renderizan como texto formateado.

Con markdown podemos, por ejemplo, introducir headers,
# Header1
## Header2
### Header3

O podemos introducir algún [link](https://www.google.com),

Incluso, podemos poner alguna imagen,

![A pretty tiger](https://upload.wikimedia.org/wikipedia/commons/5/56/Tiger.50.jpg)

O alguna ecuación en LaTex,

\begin{equation}
\bar{x} = \frac{\sum_{i=1}^n x_i}{n}
\end{equation}

# Tipos de datos en Python

Al trabajar con Python, es importante saber que cada valor tiene un tipo de dato asociado. El tipo de dato define qué clase de información estás manejando (por ejemplo, un número, una palabra, una lista de elementos, etc.) y determina qué operaciones puedes realizar con ese valor.

Python es un lenguaje de tipado dinámico, lo que significa que no necesitas declarar el tipo de una variable explícitamente: Python lo detecta automáticamente en función del valor que le asignas. Aun así, es fundamental que como programador tengas claro qué tipo de dato estás usando para evitar errores y aprovechar al máximo las funcionalidades del lenguaje.

En esta sección exploraremos los tipos de datos más comunes en Python, cómo identificarlos y cómo trabajar con ellos. Comenzaremos con los tipos numéricos.

## Funciones personalizadas en Python

En Python, además de usar funciones integradas (`print()`, `len()`, `type()`, etc.), puedes definir tus propias funciones según las necesidades de tu programa. A estas se les llama **funciones personalizadas**.

Crear tus propias funciones te permite encapsular una idea o un procedimiento, hacerlo reutilizable, y mantener tu código más ordenado y claro.

### Estructura básica

```python
def nombre_funcion(parámetros_opcionales):
    # código a ejecutarse
    return valor_opcional
```

* Se define una función con `def`, cuyo nombre debería aportar contexto.
* Dentro de un paréntesis se escriben los parametros de entrada (igual, que aporten contexto, si es posible). Estos parametros pueden tener asignado un valor por default, si es necesario.
* Escribir la lógica de la función dentro de este bloque indentado. Donde termine esta indentación es donde termina la función
* Las variables (y parámetros) declaradas dentro de la función sólo serán llamadas y guardadas dentro de la misma.
* utilizar `return` para especificar un valor de salida
* Al llamar a una función uno puede pasar los parámetros asignados de manera secuencial o explicitamente poniendo los nombres.

````python
# una funcion que cambie una temperatura en celsius a fahrenheit

def celsius_to_fahrenheit(celsius_temp = 0): # si uno no asigna valor, por defecto el input es cero
    result = (celsius_temp * 1.8) + 32
    return result, celsius_temp
````

## Estructuras de control en Python

Las **estructuras de control** permiten modificar el flujo de ejecución de un programa. Gracias a ellas, puedes tomar decisiones, repetir acciones múltiples veces y reaccionar ante diferentes condiciones.

En Python, las estructuras de control más comunes son:

- Condicionales (`if`, `elif`, `else`)
- Bucles (`for`, `while`)
- Sentencias de control de flujo (`break`, `continue`, `pass`)

Estas herramientas son fundamentales para que un programa deje de ser simplemente una lista de instrucciones lineales, y comience a responder dinámicamente a diferentes situaciones.

A continuación veremos una introducción a cada una de ellas.

### Loops o bucles
De forma frecuente, uno necesita repetir instrucciones una y otra vez. Sin embargo, en programación, si uno tiene que realizar algo 100 veces, no es necesario escribir 100 líneas de código idénticas. En Python, uno puede iterar sobre una secuencia y realizar operaciones sobre todos los elementos de una lista, tupla, string, o diccionario, por ejemplo.

* En python, se utiliza `for` y `while` para referirnos a loops.
* Todo el código indentado dentro de este `for` se ejecuta por cada iteración realizada.


````python
my_list = [10,20,30,40,50]
````

````python
len(my_list)
````

````python
range(0,5)
````

````python
range(0,len(my_list))
````

````python
for i in range(0,len(my_list)):
    print(my_list[i]+10)

# en este caso, la variable i itera sobre un rango entre 0 y 5 (no se incluye este ultimo indice 5)
# siendo que en la primera iteracion i toma el valor 0
# en la segunda iteracion i toma el valor 1
# y así hasta que i llega a tomar el valor 4
# Este índice i en este caso es utilizado para referirse al índice de una lista
````

````python
for i in range(0,5):
    print(i)
````

````python
subjects = ['Matemáticas', 'Estadística', 'Programación']

for i in range(0,len(subjects)):
    print(subjects[i])
````

````python
#también se puede hacer una iteración sobre cada elemento de una lista, sin tener que referir indice
x=1

for element in subjects:
    print(element)
````

## Diccionarios

* Es una colección **mutable** de pares clave-valor, sin orden particular
* Una **clave** es un identificador ÚNICO, acompañado de un valor.
* Un **valor** es asignado a una llave, pero éstos no son necesariamente únicos.
* Se utilizan diccionario para referirse a asociaciones, para mapear o conectar dos valores juntos.
* A diferencia de las listas, éstas se declaran con llaves {}
* Cada par clave-valor es separado por coma, y una clave es separada por `:` de sus valores.


````python
scores = {
    "Juan Perez": 98,
    "Ernesto Flores": 78,
    "Luisa Rodríguez": 100,
    "Susana Espinoza": 80
}
````

````python
type(scores)
````

````python
len(scores) # la longitud de un diccionario se refiere al numero de pares clave-valor
````

````python
# Como es mutable, podemos reasignar sus valores
# Por ejemplo, Susana se le suben 10 puntos
scores['Susana Espinoza'] = 90
print(scores)
````

````python
scores.pop('Juan Perez') # con el método pop podemos eliminar pares clave-valor
````

## Tuplas
Las tuplas son una estructura de datos que permite almacenar múltiples elementos en una sola variable. Son similares a las listas pero a diferencia de estas, **las tuplas son inmutables**, es decir, que una vez que se crean, no se pueden modificar (agregar, eliminar o cambiar elementos). Los paréntesis son utilizados para definir una tupla.

````python
tupla = (10, 20, 30, 40)
print(tupla[0])  # Salida: 10
print(tupla[2])  # Salida: 30
````

````python
# Tupla con números
tupla_numeros = (1, 2, 3, 4)

# Tupla con diferentes tipos de datos
tupla_mixta = (1, "Hola", 3.14, True)

# Tupla vacía
tupla_vacia = ()

# Tupla con un solo elemento (agregar una coma al final)
tupla_un_elemento = (10,)
````

````python
tupla = (5, 10, 15, 20)
print(tupla[1])  # Salida: 10
print(tupla[-1]) # Salida: 20 (último elemento)
````

````python
tupla = (1, 2, 3, 4, 5)
subtupla = tupla[1:4]  # Desde el índice 1 hasta el 3
print(subtupla)  # Salida: (2, 3, 4)
````

````python
tupla = (10, 20, 30, 40)
print(20 in tupla)  # Salida: True
print(50 in tupla)  # Salida: False
````
